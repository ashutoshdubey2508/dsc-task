```
#import dependences
import pandas as pd
import numpy as np
#from sklearn import linear_model
#from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
```

```
#load the boston housing dataset and transform in dataframe
data=pd.read_csv("Boston.csv")
data.head()
```

```
data.dtypes #data types of columns
```

```
#checking missing values
data.isnull().sum()
```

```
#drop ID column
data=data.drop(data.columns[0],axis=1)
print(data)
```

```
data.describe()
```

```
y=data['medv']
X=data.drop('medv',axis=1)
```

```
from sklearn.linear_model import LinearRegression

```

```
reg=LinearRegression()
```

```
# TODO: Import 'train_test_split'
from sklearn.model_selection import train_test_split

# TODO: Shuffle and split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```
reg.fit(X_train, y_train)
```

```
reg.coef_
```

```
a=reg.predict(X_test)
a
```

```
y_test
```

```
#mean squared error
np.mean((a-y_test)**2)
```

```
#graphical analysis
import matplotlib.pyplot as plt 
plt.plot(X_test,a,color='b',marker='o',markersize=3)
plt.legend(["actual prize"])
plt.title("Boston Housing Prize Prediction")
plt.xlabel("Input Features")
plt.ylabel("Prizes")
```

```

plt.plot(X_test,y_test,color='g',marker='o',markersize=3)
plt.legend(["predicted prize "])
plt.title("Boston Housing Prize Prediction")
plt.xlabel("Input Features")
plt.ylabel("Prizes")
```

```
plt.plot(X_test,a,color='b',marker='o',markersize=3)
plt.plot(X_test,y_test,color='g',marker='o',markersize=3)
plt.legend(["actual prize "],loc=0)
plt.title("Boston Housing Prize Prediction")
plt.xlabel("Input Features")
plt.ylabel("Prizes")
```
